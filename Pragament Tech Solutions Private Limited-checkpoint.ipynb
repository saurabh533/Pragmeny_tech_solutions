```python
import cv2
import numpy as np
import pytesseract
# Import required libraries useful for our project
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
#Set your path of tesseract execuitable according to your location of tesseract
```

```python
# AS we have black-white (greyscale image) no need to convert image into grey scale but you can 
# do it in case if user gives RGB input
img1 = cv2.imread('question_paper.jpg', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('filled_answer_sheet.jpg', cv2.IMREAD_GRAYSCALE)

if img1 is None or img2 is None:
        raise ValueError("IMAGES NOT READABLE")
        
#Compare image pixcel in both images
difference = cv2.absdiff(img1, img2)

#Extracting text from our image
# Use pytesseract to extract text from the image
text = pytesseract.image_to_string(difference)
```

```python
# Display the grayscale image with pixel values ranging from 0 to 255
cv2.imshow("Grayscale Image", difference)
cv2.waitKey(0)
cv2.destroyAllWindows()
#Please press any random button to stop display of image and process code further
```

```python
# Print the extracted text from the image in a python list object
print(text.split())
```
